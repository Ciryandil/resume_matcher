```python
import os
from pdfminer.high_level import extract_text

def find_sections(text):
  # Define the regular expression pattern
  pattern = r'\n([A-Za-z]+\s?[A-Za-z]+(?=:?\n))'

  # Find all matching substrings and their indices and lengths
  matches = [(match.group(0), match.start(), len(match.group(0))) for match in re.finditer(pattern, text)]
  all_matches = []
  # Print the matches, their start indices, and lengths
  for match, start_idx, length in matches:
      all_matches.append(match[1:])

  return all_matches


data_dir = '/content/gdrive/MyDrive/Resume_Matcher/archive/data/data'

sections_dir = {}
for dir in os.listdir(data_dir):
  dir_path = os.path.join(data_dir, dir)
  for resume in os.listdir(dir_path):
    resume_path = os.path.join(dir_path, resume)
    resume_text = extract_text(resume_path)
    sections_found = find_sections(resume_text)
    for section in sections_found:
      if section in sections_dir:
        sections_dir[section] += 1
      else:
        sections_dir[section] = 1

print(sections_dir)
```

```python
import os
from pdfminer.high_level import extract_text
import json

parsed_resumes = {}

resume_headers = json.load(open('resume_headers.json'))



def extract_sections(text):
  headers_dict= {}
  sections_dict = {}
  for header in resume_headers:
    header_1 = '\n' + header + '\n'
    header_2 = '\n' + header + ':'
    header_3 = '\n' + header + ':\n'
    if text.find(header_1) != -1:
      headers_dict[header] = {'length': len(header_1), 'index': text.find(header_1)}
    elif text.find(header_2) != -1:
      headers_dict[header] = {'length': len(header_2), 'index': text.find(header_2)}
    elif text.find(header_3) != -1:
      headers_dict[header] = {'length': len(header_3), 'index': text.find(header_3)}

  sorted_headers_dict = dict(sorted(headers_dict.items(), key=lambda item: item[1]['index']))
  # print(sorted_headers_dict)
  # print(headers_dict)

  headers = list(sorted_headers_dict.keys())
  # print(headers)
  for i in range(len(headers)):
    header = headers[i]
    if i < len(headers) - 1:
      next_header = headers[i+1]
      # print(header)
      # print(next_header)
      sections_dict[header] = text[sorted_headers_dict[header]['index']+sorted_headers_dict[header]['length']:
                                   sorted_headers_dict[next_header]['index']]
    else:
      sections_dict[header] = text[sorted_headers_dict[header]['index']+sorted_headers_dict[header]['length']:]

  return sections_dict





```

```python
def parse_job_desc_skills(job_desc):
  split_skills = [item for item in job_desc['Required Skills'].split('.')]
  #print(split_skills)
  skill_set = []
  for item in split_skills:
    skill_set.extend(item.split(','))
  #print(skill_set)
  skill_set = [item.strip() for item in skill_set if not item == '']
  return skill_set
def parse_resume_skills(resume):
  skill_set = []
  if resume['Skills'].find(',') == -1:
    skill_set = resume['Skills'].split('\n')
  else:
    skill_set = resume['Skills'].split(',')

  skill_set = [item.strip() for item in skill_set if item != '']

  return skill_set
```

```python
import torch
import torch.nn.functional as F

def find_max_similarity(list1, list2):
    # Convert the lists of vectors to PyTorch tensors
    list1 = torch.tensor(list1)
    list2 = torch.tensor(list2)

    # Calculate cosine similarity between all pairs of vectors
    similarity_matrix = F.cosine_similarity(list1.unsqueeze(1), list2.unsqueeze(0), dim=2)


    max_similarity_values = torch.max(similarity_matrix, dim=1).values


    res = torch.mean(max_similarity_values, dim=0)
    return res.detach().numpy()


# # Example usage:
# # Assuming list1 and list2 are lists of PyTorch vectors (tensors)
# list1 = [torch.tensor([1.0, 2.0, 3.0]), torch.tensor([4.0, 5.0, 6.0])]
# list2 = [torch.tensor([7.0, 8.0, 9.0]), torch.tensor([10.0, 11.0, 12.0])]

# max_similarity = find_max_similarity(embeddings_job, embeddings_resume)
# print(max_similarity)



#print(torch.mean(res))

```

```python
import re
def parse_section(section_text, section_key):
  section_items = None
  if section_text.find('.') == -1 and section_text.find(',') == -1:
    section_items = section_text.split('\n')

  elif (section_text.find('.') == -1 and (section_text.find(',') != -1 or section_text.find(';') != -1)) or 'Skill' in section_key:
    section_items = re.split(r'[,;]', section_text)
  elif section_text.find('.') != -1:
    section_items = section_text.split('.')
  else:
    section_items = [section_text]

  section_items = [re.sub('\n+', ' ', item.strip()) for item in section_items if item.strip() != '' and item.strip() != '\n' and item.strip() != '\t']

  return section_items
```

```python
import json
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/sentence-t5-base')
result_dict = {}
parsed_jd = {}
encoded_jd = {}
parsed_resume = {}
encoded_resume = {}

for i in tqdm(range(job_descs.shape[0])):
  job_description = json.loads(job_descs['model_response'][i])

  for jd_key in list(job_description.keys()):
    if not i in parsed_jd.keys():
      parsed_jd[i] = {}
      encoded_jd[i] = {}
    if isinstance(job_description[jd_key], list):
      parsed_jd[i][jd_key] = job_description[jd_key]
    else:
      parsed_jd[i][jd_key] = parse_section(job_description[jd_key], jd_key)
    encoded_jd[i][jd_key] = model.encode(parsed_jd[i][jd_key])

for resume_i in tqdm(list(resumes.keys())):
  resume = resumes[resume_i]

  for r_key in list(resume.keys()):
    if not resume_i in parsed_resume.keys():
      parsed_resume[resume_i] = {}
      encoded_resume[resume_i] = {}
    parsed_resume[resume_i][r_key] = parse_section(resume[r_key], r_key)
    encoded_resume[resume_i][r_key] = model.encode(parsed_resume[resume_i][r_key])
```

```python
result_dict = {}
for jd_key in tqdm(list(encoded_jd.keys())[:100]):
  for resume_i in list(encoded_resume.keys()):
    sim_arr = []
    for key in list(encoded_jd[jd_key].keys()):
      if len(encoded_jd[jd_key][key]) == 0:
        continue
      if key not in list(mapping_dict.keys()):
        continue
      corr_keys = mapping_dict[key]
      resume_encoding_corr = []
      for r_key in corr_keys:
        if r_key in list(encoded_resume[resume_i].keys()):
          resume_encoding_corr.extend(encoded_resume[resume_i][r_key])

      if len(resume_encoding_corr) == 0:
        sim_arr.append(0)
      else:
        # print(list(encoded_jd.keys()))
        sim_arr.append(find_max_similarity(encoded_jd[jd_key][key], resume_encoding_corr))

    if not jd_key in result_dict:
      result_dict[jd_key] = {}
      result_dict[jd_key][resume_i] = sum(sim_arr) / len(sim_arr)
    else:
      result_dict[jd_key][resume_i] = sum(sim_arr) / len(sim_arr)
    #print(result_dict)

```
